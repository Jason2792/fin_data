```python
import numpy as np
import scipy.stats as si
import csv
import matplotlib.pyplot as plt 
import pandas as pd 
file = open('mydata1.csv','w')
file = csv.writer(file)
file.writerow(['K','BS_call','BS_put'])
for i in range(20,101,1): 
    print(i)
    S, K, T, r, sigma = 90, i, 0.5, 0.01, 0.3
    def BS(S,K,T,r,sigma): 
        d1 = (np.log(S/K)+(r+0.5*sigma**2)*T)/(sigma*np.sqrt(T))
        d2 = (np.log(S/K)+(r-0.5*sigma**2)*T)/(sigma*np.sqrt(T))    
        call = S*si.norm.cdf(d1,0.0,1.0)-K*np.exp(-r*T)*si.norm.cdf(d2,0.0,1.0)
        put = -S*si.norm.cdf(-d1,0.0,1.0)+K*np.exp(-r*T)*si.norm.cdf(-d2,0.0,1.0)
        return call,put
    BS_call, BS_put= BS(S,K,T,r,sigma)
    print ('S,K,T,r,sigma=',S,K,T,r,sigma)
    print ('BS_call=',round(BS_call,4),', BS_put=', round(BS_put,4))
    file.writerow([i,round(BS_call,4),round(BS_put,4)])
file = pd.read_csv('mydata1.csv')  
x_axis = file['K'] 
y_axis1 = file['BS_put'] 
y_axis2 = file['BS_call']
plt.plot(x_axis, y_axis1)
plt.plot(x_axis, y_axis2)
plt.show()   
```
